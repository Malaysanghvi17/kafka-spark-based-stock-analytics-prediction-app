```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.sql.functions import lit
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, DoubleType
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
# Create a Spark session
spark = SparkSession.builder.appName("StockPriceProcessing").getOrCreate()

# Define the schema for the CSV data
schema = StructType([
    StructField("Date", StringType(), True),
    StructField("Open", DoubleType(), True),
    StructField("High", DoubleType(), True),
    StructField("Low", DoubleType(), True),
    StructField("Close", DoubleType(), True),
    StructField("Volume", DoubleType(), True)
])
```

```python
def load_csv_spark(stock_name):
    # Read the CSV data into a Spark DataFrame
    data_dir = "data"
    
    stock_file = stock_name + ".csv"
    # stock_files = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith(".csv")]
    stock_file = [os.path.join(data_dir, stock_file)]
    combined_df = spark.read.csv(stock_file, header=True, inferSchema=True)
    return combined_df
combined_df = load_csv_spark("AAPL")
combined_df.show(5)
```

```python
# Register the DataFrame as a temporary view
combined_df.createOrReplaceTempView("stock")

# Run SQL queries on the DataFrame
time = spark.sql("SELECT Date FROM stock WHERE Open IS NOT NULL")
open_prices = spark.sql("SELECT Open FROM stock WHERE Open IS NOT NULL")
high_prices = spark.sql("SELECT High FROM stock WHERE Open IS NOT NULL")
low_prices = spark.sql("SELECT Low FROM stock WHERE Open IS NOT NULL")
close_prices = spark.sql("SELECT Close FROM stock WHERE Open IS NOT NULL")
```

```python
time_df = time.toPandas()
open_prices_df = open_prices.toPandas()
high_prices_df = high_prices.toPandas()
low_prices_df = low_prices.toPandas()
close_prices_df = close_prices.toPandas()
```

```python


def plot_time_series(time_df, prices_df, title):
    plt.figure(figsize=(12, 6))
    for column in prices_df.columns:
        plt.plot(time_df['Date'], prices_df[column], label=column)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_bar_chart(data_df, x_column, y_column, title):
    plt.figure(figsize=(10, 6))
    plt.bar(data_df[x_column], data_df[y_column], color='skyblue')
    plt.title(title)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def plot_box_plot(data_df, column, title):
    plt.figure(figsize=(8, 6))
    plt.boxplot(data_df[column])
    plt.title(title)
    plt.ylabel(column)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_scatter_plot(data_df, x_column, y_column, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(data_df[x_column], data_df[y_column], color='green', alpha=0.5)
    plt.title(title)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Usage example:
# Plot Time Series
# plot_time_series(time_df, open_prices_df, 'Open Prices Over Time')

# Plot Bar Chart
plot_bar_chart(time_df, 'Date', 'Open', 'Open Prices')

# Plot Box Plot
# plot_box_plot(open_prices_df, 'Open', 'Box Plot of Open Prices')

# Plot Scatter Plot
# plot_scatter_plot(close_prices_df, 'Close', 'Volume', 'Close Prices vs Volume')
```

```python

def plot_time_series_seaborn(time_df, prices_df, title, max_data_points=1000):
    plt.figure(figsize=(12, 6))
    step = max(1, len(time_df) // max_data_points)
    for column in prices_df.columns:
        sns.lineplot(x=time_df['Date'][::step], y=prices_df[column][::step], label=column)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_bar_chart_seaborn(data_df, x_column, y_column, title, max_data_points=1000):
    plt.figure(figsize=(10, 6))
    data_df = data_df.iloc[:max_data_points]
    sns.barplot(x=data_df[x_column], y=data_df[y_column], color='skyblue')
    plt.title(title)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Usage example:
# Plot Time Series with Seaborn
plot_time_series_seaborn(time_df, open_prices_df, 'Open Prices Over Time', max_data_points=1000)

# Plot Bar Chart with Seaborn
# plot_bar_chart_seaborn(time_df, 'Date', 'open_prices_df', 'Open Prices', max_data_points=1000)
```
